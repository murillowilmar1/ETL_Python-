# E - Extracción de datos 

```python

#### Conexion con la base de datos SQL Server 

server = '******'
database = '*****'
username = '*****'
password = '*****'

conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=' + server + ';DATABASE=' + database + ';UID=' + username + ';PWD=' + password)


cursor = conn.cursor()

print("Conexión existosa......")

```
